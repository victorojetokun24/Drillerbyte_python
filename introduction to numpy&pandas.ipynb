# introduction to matplotlib for visulaization

```python
# define x & y value
x = new_pf['XEL UN Equity']
y = new_pf['AA UN Equity']
```

```python
plt.plot(x,y)
plt.show()
```

```python
# types of plot graph. (1) as Scatter
plt.scatter(x,y)
plt.xlabel('XEL')
plt.ylabel('AA')
plt.show()
```

```python
 # histogram plot
plt.hist(y)
plt.xlabel('XEL')
plt.ylabel('AA')
plt.show()
```

```python
# bar chat
plt.bar(x,y)
plt.ylabel('AA')
plt.show()
```

```python
#pie chart
plt.pie(y, labels=new_pf['AA UN Equity'])
plt.show()
```
